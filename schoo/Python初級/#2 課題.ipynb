```python
from typing import Optional


def ask(exercise_dict: dict) -> Optional[tuple[int, int, int]]:
    ask_exercise = f"実施した運動タイプはどれですか？ [{', '.join(exercise_dict.keys())}]"
    ask_weight = f"体重を入力してください"
    ask_time = f"運動時間を入力してください"

    # 運動タイプの入力
    try:
        input_exercise = input(ask_exercise)
        if input_exercise not in exercise_dict.keys():
            raise ValueError("入力した運動タイプは選択肢に存在しません。")
    except ValueError as e:
        print(f"入力エラー: {e}")
        return None

    # 体重の入力
    try:
        input_weight = input(ask_weight)
        input_weight_num = int(input_weight)
    except ValueError as e:
        print(f"入力エラー: 体重は数値で入力してください。")
        return None

    # 運動時間の入力
    try:
        input_time = input(ask_time)
        input_time_num = int(input_time)
    except ValueError as e:
        print(f"入力エラー: 運動時間は数値で入力してください。")
        return None

    return (exercise_dict[input_exercise], input_weight_num, input_time_num)

def calc_calorie(mets: int, weight: int, time:int) -> None:
    """運動、体重、時間からカロリーを計算する"""
    result = mets * weight * time * 1.05 / 60
    print(f"消費カロリーは{result}kcalです")
```

```python
def main():
    exercise_dict = {"running": 9, "walking": 3, "cycling": 5}

    mets, weight, time = ask(exercise_dict)
    calc_calorie(mets, weight, time)

main()
```
