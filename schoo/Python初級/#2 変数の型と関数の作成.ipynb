### 別の関数例1

```python
# 徒歩の消費カロリー
# 3 × 体重 × 運動時間(分数) × 1.05 ÷ 60
a = 60
b = 50
```

```python
3 * a * b * 1.05 / 60
```

```python
def calories_walking(weight, minutes):
    """ウォーキングのカロリー計算"""
    return 3 * weight * minutes * 1.05 / 60
```

```python
calories_walking(60, 30)
```

```python
calories_walking(80, 60)
```

```python
calories_walking(70, 90)
```

```python
calories_walking(50, 15)
```

### 別の関数例2

```python
def calories_running(weight, minutes):
    """ランニングのカロリー計算"""
    return 9 * weight * minutes * 1.05 / 60
```

```python
calories_running(60, 30)
```

```python
calories_running(80, 60)
```

```python
calories_running(70, 90)
```

```python
calories_running(50, 15)
```
