### ウォーキングの消費カロリーを計算
ウォーキングの消費カロリーは<br>
**3 × 体重(kg) × 運動時間(分数) × 1.05 ÷ 60 で計算されるらしい...**<br>
例えば、60kgの人が50分運動したら...

```python
3 * 60 * 50 * 1.05 / 60
```

```python
# 体重: 60kg, 運動時間: 90分
3 * 60 * 90 * 1.05 / 60
```

```python
# 体重: 80kg, 運動時間: 40分
3 * 80 * 40 * 1.05 / 60
```

### 変数を使って計算

```python
# 3 × 体重 × 運動時間(分数) × 1.05 ÷ 60
weight = 60
minutes = 50

3 * weight * minutes * 1.05 / 60
```

```python
weight = 60
minutes = 90

3 * weight * minutes * 1.05 / 60
```
