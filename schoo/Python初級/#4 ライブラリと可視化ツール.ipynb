## モジュールの使用方法


**モジュールの使用方法**\
import モジュール名\
モジュール名.関数名(引数)

```Python
import math
math.sqrt(10)
math.sin(100)
```

**読み込む際に別名をつける**\
import モジュール名 as 別名

```Python
import math as mm
mm.sqrt(10)
mm.sin(100)
```

````python
from matplotlib import pyplot as plt
import seaborn as sns
plt.style.use("ggplot")
%matplotlib inline

from sklearn.datasets import load_iris
import pandas as pd
````

````python
iris = load_iris()

df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
df
````

````python
# seabornではjointplot
sns.jointplot(x="sepal length (cm)", y="petal length (cm)", data=df)
````

````python
sns.histplot(df["sepal length (cm)"])
````

````python
x = [1, 2, 3, 4]
y = [2, 5, 7, 9]

sns.lineplot(x=x, y=y)
````

````python
sns.boxplot(x="target", y="sepal length (cm)", data=df)
````

````python
corr = df.corr()
sns.heatmap(corr)
````
