## モジュールの使用方法


**モジュールの使用方法**\
import モジュール名\
モジュール名.関数名(引数)

```Python
import math
math.sqrt(10)
math.sin(100)
```

**読み込む際に別名をつける**\
import モジュール名 as 別名

```Python
import math as mm
mm.sqrt(10)
mm.sin(100)
```

## randomモジュール
